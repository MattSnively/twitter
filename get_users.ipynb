## Retrieving users data

### search_users()

If you completed the labs in last week's module, then you already know it's possible to search for tweets using Twitter's REST API. A lesser known Twitter API end-point, [`users/search`](https://developer.twitter.com/en/docs/accounts-and-users/follow-search-get-users/api-reference/get-users-search), actually makes it possible to search for Twitter users as well. The `users/search` API returns up to 1,000 users by keyword searching user bios.

To access the `users/search` API with rtweet, use the `search_users()` function. For a description of the parameters accepted by `search_users()`, execute the following code block:

```r
?search_users
```

The example code below uses `search_users()` to search through account (user) descriptions for the keywords bitcoin and cryptocurrency. Because the two terms are separated by `OR`, the first 1,0000 users with descriptions that include either (or both) keywords will be returned. I've included `verbose = FALSE` to suppress the processing-notification message (an aesthetically driven decision; the message is a little clumsy and shows up with a red background in jupyterhub).

```r
## search for 1000 users using keywords bitcoin or cryptocurrency
cc <- search_users(q = "bitcoin OR cryptocurrency", n = 1000, verbose = FALSE)

## a tibble (data frame) where each observation (row) is a different user
head(cc)
```

The returned data object `cc` also contains a "tweets" attribute, which can be accessed via `tweets_data()`. This tweets data consists of the most recent [single] tweet posted by each user in `cc`.

```r
## a tibble (data frame) where each observation (row) is a tweet
cc %>%
    tweets_data() %>%
    head()
```

### lookup_users()

If you already have a vector of user IDs or screen names, you can also lookup data on those users with the `lookup_users()` function. For a description of the parameters accepted by `search_users()`, execute the following code block:

```r
?lookup_users
```

For a quick example of how `lookup_users()` works, the code below looks up users data on four different news organization accounts.

```r
## use user_id or screen_name vector to lookup users data
users <- c("latimes", "cnn", "nytimes", "foxnews")
usrs <- lookup_users(users)
```

### Users data

Users-level variables include the following variables
- **user_id**: unique identifiers
- **name**: full name listed by user
- **screen name**: twitter user handle
- **description**: profile bio statement
- **protected**: logical, incidating whether user's profile is set to private
- **followers_count**: number of followers
- **friends_count**: number of friends
- **listed_count**: number of lists user appears on
- **statuses_count**: number of tweets posted by user
- **favourites_count**: number of tweets favorited by user
- **account_created_at**: account creation/origin date
- **verified**: logical indicating whether the account is verified
- **profile_url**: URL listed by user
- **profile_expanded_url**: full (expanded) URL listed by user
- **account_lang**: language associated with the user's profile
- **profile_banner_url**: URL to user profile banner image
- **profile_background_url**: URL to user profile background image
- **profile_image_url**: URL to user profile image
