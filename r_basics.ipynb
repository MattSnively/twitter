
**9. Clean and convert all appropriate columns to numeric.**

```r
## your code goes here
## identify the column(s) that should probably be considered integer or numeric
nums <- c("rank", "cultivated_land_km2", "cultivated_land_percent", "arable_land_km2", "arable_land_percent", "permanent_crops_km2", "permanent_crops_percent", "other_lands_km2", "other_lands_percent", "total_area_km2", "date")

## remove everything that's not a number, minus sign, or decimal and convert to numeric
landuse_df[nums] <- lapply(landuse_df[nums], function(x)
  as.numeric(stringr::str_replace_all(x, "[^[:digit:]\\-\\.]", "")))

## set rank var as integer
landuse_df$rank <- as.integer(landuse_df$rank)

## preview data
landuse_df
```
